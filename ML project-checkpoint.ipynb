# Feature extraction using PCA

```python
# Separating out the features
X = df.iloc[:, 7: ]
# Separating out the gas
y = df.gas
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(X)
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
 
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```

```python
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['gas']]], axis = 1)
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1 - {0}%'.format(per_var[0]), fontsize = 15)
ax.set_ylabel('Principal Component 2 - {0}%'.format(per_var[1]), fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
gases = ['mixture', 'ethanol', 'acetone','air']
colors = ['r', 'g', 'b', 'c']
for gas, color in zip(gases,colors):
    indicesToKeep = finalDf['gas'] == gas
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(gases)
ax.grid()
```

# Decision Tree Classifier

```python
y2 = pd.factorize(df.gas)[0]
y2
```

```python
def print_tree(estimator, features, class_names=None, filled=True):
    tree = estimator
    names = features
    color = filled
    classn = class_names
    dot_data = StringIO()
    export_graphviz(estimator, out_file=dot_data, feature_names=features, class_names=classn, filled=filled)
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    return(graph)
```

```python
from sklearn.preprocessing import StandardScaler
# Dividing X, y into train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y2,test_size=0.3, random_state = 0)
# Standardizing the features
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
clf = DecisionTreeClassifier(max_depth=6)
clf.fit(X_train, y_train)
```

```python
# Plot the tree
import pydotplus
from IPython.display import Image
graph = print_tree(clf, features=X.columns.values , class_names=['air', 'ethanol','acetone','mixture'])
Image(graph.create_png())
```

```python
#fitting the model
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
pred
```

```python
#confusion matrix
cm = pd.DataFrame(confusion_matrix(y_test, pred).T, index=['air', 'ethanol','acetone','mixture'], columns=['air', 'ethanol','acetone','mixture'])
cm.index.name = 'Predicted'
cm.columns.name = 'Gas'
cm
```

```python
# Precision of the model using test data is 70%
print(classification_report(y_test, pred))
```

```python
from sklearn.metrics import accuracy_score
print ("Accuracy is",  accuracy_score(y_test,pred)*100)
```

```python
print ('Test MSE: ', mean_squared_error(y_test, clf.predict(X_test)))
```

```python
print ('Root_Mean_Squared_Error : ', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

# Support Vector Machine (SVM)
